```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.chrome.service import Service
import openpyxl
```

```python
s=Service("D:\chromedrive\chromedriver.exe")
driver=webdriver.Chrome(service=s)
```

```python
import pandas as pd
```

```python
data = pd.read_csv('newnewbse.csv')
```

```python
data
```

```python
driver.get('https://www.zaubacorp.com/')
driver.maximize_window()
```

```python
data['Description'][3999:4001]
```

```python
temp=[]
for i in range(4000,4077):
    print(data['CIN'][i])
    inputElement = driver.find_element_by_id('searchid')
    if type(data['CIN'][i])!=float:
        try:
            inputElement.send_keys(data['CIN'][i])
            inputElement.send_keys(Keys.ENTER)
            ele=driver.find_elements(By.XPATH,value='//*[@id="block-system-main"]/div[2]/div[1]')
            x=ele[0].text.split('Company Details')[0]
            temp.append(x)
        except:
            temp.append("")
    else:
        temp.append("")
        continue
```

```python
len(temp)
```

```python
data['Description'][4000:4077]=temp
```

```python
data.to_csv("newbse.csv",index=False)
```
